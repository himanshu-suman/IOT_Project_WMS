```python
# import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
plt.style.use("seaborn")
```

```python
# import fiveyearsweatherdata_kolkata.csv file
df = pd.read_csv('fiveyearsweatherdata_kolkata.csv')
df.head()
```

### Dataset

Dataset taken from Kaggle

```python
# print number of rows and columns
print('Number of rows and columns in the dataset are: ', df.shape)
```

```python
# list all column names
print('Column names are: ', df.columns)
```

```python
# describe the dataset
df.describe()
```

```python
# check for null values
df.isnull().sum()
```

```python
# drop Wind Gust, Wind Chill, Info because they have too many null values
df.drop(['Wind Gust', 'Wind Chill', 'Info'], axis=1, inplace=True)
```

```python
# find minimum and maximum values of Heat Index
print('Minimum value of Heat Index is: ', df['Heat Index'].min())
print('Maximum value of Heat Index is: ', df['Heat Index'].max())
```

```python
# replace null values of Heat Index with mean of Heat Index
df['Heat Index'].fillna(df['Heat Index'].mean(), inplace=True)
```

```python
# check again for null values
df.isnull().sum()
```

```python
# drop null values
df.dropna(inplace=True)
```

```python
# check shape of the dataset
df.shape
```

```python
# check for duplicates
df.duplicated().sum()
```

### Finding Insights

```python
df.head()
```

```python
# find the average temperature of each day
df['Day'] = pd.DatetimeIndex(df['Date time']).day

# find average temperature of each day
df.groupby('Day')['Temperature'].mean()

# plot average temperature of each day
plt.figure(figsize=(10, 5))
sns.barplot(x=df.groupby('Day')['Temperature'].mean().index, y=df.groupby('Day')['Temperature'].mean().values)
plt.xlabel('Day')
plt.ylabel('Average Temperature')
plt.title('Average Temperature of Each Day')
plt.show()
```

```python
# find the average temperature of each month
df['Month'] = pd.DatetimeIndex(df['Date time']).month

# find average temperature of each month
df.groupby('Month')['Temperature'].mean()

# plot average temperature of each month
plt.figure(figsize=(10, 5))
sns.barplot(x=df.groupby('Month')['Temperature'].mean().index, y=df.groupby('Month')['Temperature'].mean().values)
plt.xlabel('Month')
plt.ylabel('Average Temperature')
plt.title('Average Temperature of Each Month')
plt.show()
```

```python
# find the average temperature of each year
df['Year'] = pd.DatetimeIndex(df['Date time']).year

# find average temperature of each year
df.groupby('Year')['Temperature'].mean()

# plot average temperature of each year
plt.figure(figsize=(10, 5))
sns.barplot(x=df.groupby('Year')['Temperature'].mean().index, y=df.groupby('Year')['Temperature'].mean().values)
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.title('Average Temperature of Each Year')
plt.show()
```

```python
# plot Temperature vs Date time using line plot
plt.figure(figsize=(20, 5))
sns.lineplot(x=df['Date time'], y=df['Temperature'])
plt.xlabel('Date time')
plt.ylabel('Temperature')
plt.title('Temperature vs Date time')
plt.show()
```

```python
# find date time when temperature was maximum and minimum
print('Date time when temperature was maximum is: ', df[df['Temperature'] == df['Temperature'].max()]['Date time'].values[0])
print('Date time when temperature was minimum is: ', df[df['Temperature'] == df['Temperature'].min()]['Date time'].values[0])
```

```python
# plot dew point vs date time using line plot
plt.figure(figsize=(20, 5))
sns.lineplot(x=df['Date time'], y=df['Dew Point'])
plt.xlabel('Date time')
plt.ylabel('Dew Point')
plt.title('Dew Point vs Date time')
plt.show()
```

```python
# find date time when dew point was maximum and minimum
print('Date time when dew point was maximum is: ', df[df['Dew Point'] == df['Dew Point'].max()]['Date time'].values[0])
print('Date time when dew point was minimum is: ', df[df['Dew Point'] == df['Dew Point'].min()]['Date time'].values[0])
```

```python
# plot relative humidity vs date time using line plot
plt.figure(figsize=(20, 5))
sns.lineplot(x=df['Date time'], y=df['Relative Humidity'])
plt.xlabel('Date time')
plt.ylabel('Relative Humidity')
plt.title('Relative Humidity vs Date time')
plt.show()
```

```python
# find date time when relative humidity was maximum and minimum
print('Date time when relative humidity was maximum is: ', df[df['Relative Humidity'] == df['Relative Humidity'].max()]['Date time'].values[0])
print('Date time when relative humidity was minimum is: ', df[df['Relative Humidity'] == df['Relative Humidity'].min()]['Date time'].values[0])
```

```python
# plot wind speed vs date time using line plot
plt.figure(figsize=(20, 5))
sns.lineplot(x=df['Date time'], y=df['Wind Speed'])
plt.xlabel('Date time')
plt.ylabel('Wind Speed')
plt.title('Wind Speed vs Date time')
plt.show()
```

```python
# find date time when wind speed was maximum and minimum
print('Date time when wind speed was maximum is: ', df[df['Wind Speed'] == df['Wind Speed'].max()]['Date time'].values[0])
print('Date time when wind speed was minimum is: ', df[df['Wind Speed'] == df['Wind Speed'].min()]['Date time'].values[0])
```

```python
# create and plot a heatmap of the correlation matrix
# plot so that the heatmap is easier to read and the text is not squished together
plt.figure(figsize=(10, 10))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='Blues')
plt.show()
```

```python
# get top 5 features that are highly correlated
df.corr().abs().unstack().sort_values(ascending=False).drop_duplicates()[1:6]
```

The relation between humidity and temperature formula simply says they are inversely proportional. If temperature increases it will lead to a decrease in relative humidity, thus the air will become drier whereas when temperature decreases, the air will become wet means the relative humidity will increase.
